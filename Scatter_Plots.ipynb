##Scatter plots

Untuk membuat scatter plot sederhana, kita menggunakan perintah `sns.scatterplot` dan menentukan nilai untuk:

- sumbu x horizontal (`x = data_asuransi['bmi']`), dan
- sumbu y vertikal (`y = data_asuransi['biaya']`).

````python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
````

Kode di atas menggunakan seaborn untuk membuat scatter plot yang menampilkan hubungan antara variabel 'bmi' (Body Mass Index) dan 'charges' (biaya asuransi) dalam data asuransi yang Kita miliki.

- `sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])`: Ini adalah panggilan fungsi `scatterplot` dari pustaka seaborn (`sns`). Fungsi ini digunakan untuk membuat scatter plot. Variabel 'bmi' digunakan sebagai sumbu-x (sumbu horizontal), sementara variabel 'charges' digunakan sebagai sumbu-y (sumbu vertikal). Scatter plot ini akan menampilkan titik-titik data di mana setiap titik mewakili satu entri dalam data asuransi, dengan 'bmi' di sumbu-x dan 'charges' di sumbu-y.

Kita dapat menempatkan baris ini di markdown notebook Kita dan menjalankannya untuk menghasilkan scatter plot yang memvisualisasikan hubungan antara 'bmi' dan 'charges' dalam data asuransi .

````python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
````

Kode di atas menggunakan seaborn untuk membuat scatter plot dengan garis regresi yang menampilkan hubungan antara variabel 'bmi' (Body Mass Index) dan 'charges' (biaya asuransi) dalam data asuransi yang Kita miliki.

- `sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])`: Ini adalah panggilan fungsi `regplot` dari pustaka seaborn (`sns`). Fungsi ini digunakan untuk membuat scatter plot dengan garis regresi. Variabel 'bmi' digunakan sebagai sumbu-x (sumbu horizontal), sementara variabel 'charges' digunakan sebagai sumbu-y (sumbu vertikal). Scatter plot ini akan menampilkan titik-titik data di mana setiap titik mewakili satu entri dalam data asuransi, dengan 'bmi' di sumbu-x dan 'charges' di sumbu-y, sambil menambahkan garis regresi untuk menunjukkan tren umum antara dua variabel tersebut.

Kita dapat menempatkan baris ini di markdown notebook Kita dan menjalankannya untuk menghasilkan scatter plot dengan garis regresi yang memvisualisasikan hubungan antara 'bmi' dan 'charges' dalam data asuransi.

##Color-coded scatter plots

````python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
````

Kode di atas menggunakan seaborn untuk membuat scatter plot yang menampilkan hubungan antara variabel 'bmi' (Body Mass Index) dan 'charges' (biaya asuransi) dalam data asuransi yang dimiliki, dengan membedakan antara perokok dan bukan perokok.

- `sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])`: Ini adalah panggilan fungsi `scatterplot` dari pustaka seaborn (`sns`). Fungsi ini digunakan untuk membuat scatter plot. Variabel 'bmi' digunakan sebagai sumbu-x (sumbu horizontal), sementara variabel 'charges' digunakan sebagai sumbu-y (sumbu vertikal). Dengan menambahkan parameter `hue=insurance_data['smoker']`, kita membedakan titik-titik berdasarkan kategori 'smoker' dalam data. Ini berarti plot akan menampilkan dua warna berbeda untuk titik-titik yang mewakili perokok dan bukan perokok.

Kita dapat menempatkan baris ini di markdown notebook Anda dan menjalankannya untuk menghasilkan scatter plot yang memvisualisasikan hubungan antara 'bmi' dan 'charges' dalam data asuransi , sambil membedakan antara perokok dan bukan perokok.

````python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
````

Kode di atas menggunakan seaborn untuk membuat scatter plot dengan garis regresi yang membedakan antara perokok dan bukan perokok, serta menampilkan hubungan antara variabel 'bmi' (Body Mass Index) dan 'charges' (biaya asuransi) dalam data asuransi.

- `sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)`: Ini adalah panggilan fungsi `lmplot` dari pustaka seaborn (`sns`). Fungsi ini digunakan untuk membuat scatter plot dengan garis regresi (regression plot). Parameter `x` dan `y` digunakan untuk menentukan variabel yang akan diplot di sumbu-x dan sumbu-y. Parameter `hue` digunakan untuk membedakan titik-titik berdasarkan kategori 'smoker' dalam data. Ini berarti plot akan menampilkan dua garis regresi berbeda untuk perokok dan bukan perokok. Parameter `data` digunakan untuk menentukan DataFrame yang akan digunakan untuk plot, dalam hal ini `insurance_data`.

Kita dapat menempatkan baris ini di markdown notebook dan menjalankannya untuk menghasilkan scatter plot dengan garis regresi yang membedakan antara perokok dan bukan perokok, sambil menampilkan hubungan antara 'bmi' dan 'charges' dalam data asuransi yang kita miliki. Ini akan memberi Kita wawasan tentang bagaimana kebiasaan merokok memengaruhi biaya asuransi berdasarkan BMI.

````python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
````

Kode di atas menggunakan seaborn untuk membuat swarm plot yang menampilkan distribusi biaya asuransi ('charges') berdasarkan kategori 'smoker' dalam data asuransi.

- `sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])`: Ini adalah panggilan fungsi `swarmplot` dari pustaka seaborn (`sns`). Fungsi ini digunakan untuk membuat swarm plot. Parameter `x` digunakan untuk menentukan variabel yang akan diplot di sumbu-x, dalam hal ini, kategori 'smoker'. Parameter `y` digunakan untuk menentukan variabel yang akan diplot di sumbu-y, dalam hal ini, 'charges' (biaya asuransi).

Kita dapat menempatkan baris ini di markdown notebook dan menjalankannya untuk menghasilkan swarm plot yang memvisualisasikan distribusi biaya asuransi berdasarkan kategori 'smoker' dalam data asuransi Anda. Ini akan membantu Anda dalam memahami bagaimana biaya asuransi bervariasi antara perokok dan bukan perokok.
